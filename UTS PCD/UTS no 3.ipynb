```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Muat gambar yang telah diunggah
image_path = "wa.jpg"  # Path gambar yang telah diupload
image = cv2.imread(image_path)

# Mengubah gambar menjadi grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Meningkatkan kecerahan dengan menambahkan konstanta pada citra grayscale
brightened_image = cv2.convertScaleAbs(gray_image, alpha=1.5, beta=50)  # Meningkatkan kecerahan dengan alpha=1.5 dan beta=50

# Meningkatkan kontras gambar yang telah dipercepat kecerahannya
brightened_and_contrasted_image = cv2.convertScaleAbs(brightened_image, alpha=2, beta=50)  # Meningkatkan kontras dengan alpha=2

# Menampilkan 4 gambar (Asli, Grayscale, Diperbesar Kecerahannya, Diperbesar Kecerahannya & Kontrasnya)
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Gambar Asli
axes[0, 0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0, 0].set_title('Gambar Asli')
axes[0, 0].axis('off')

# Gambar Grayscale
axes[0, 1].imshow(gray_image, cmap='gray')
axes[0, 1].set_title('Gambar Grayscale')
axes[0, 1].axis('off')

# Gambar Grayscale yang Diperbesar Kecerahannya (alpha=1.5, beta=50)
axes[1, 0].imshow(brightened_image, cmap='gray')
axes[1, 0].set_title('Gambar Dipercerah (alpha=1.5, beta=50)')
axes[1, 0].axis('off')

# Gambar Grayscale yang Diperbesar Kecerahannya dan Diperkontras (alpha=2)
axes[1, 1].imshow(brightened_and_contrasted_image, cmap='gray')
axes[1, 1].set_title('Gambar Dipercerah & Diperkontras (alpha=2)')
axes[1, 1].axis('off')

# Menampilkan semua gambar
plt.tight_layout()
plt.show()



#Fakhliq Ainul Yaqin 202331329
```
